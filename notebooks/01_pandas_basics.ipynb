```python
import pandas as pd
print("Pandas 版本:", pd.__version__)
```

```python
#建立練習用的模擬台股資料表
data={
    "股票代號": ["2330", "2317", "2454", "1301", "2882"],
      "公司名稱": ["台積電", "鴻海", "聯發科", "台塑", "國泰金"],
      "收盤價": [895, 178, 1320, 68.5, 52.3],
      "成交量(張)": [45230, 32100, 8750, 12400, 18900],
      "產業": ["半導體", "電子代工", "半導體", "塑化", "金融"]
      }
df=pd.DataFrame(data)
df
```

```python
#學習操作panda表格之，只看前三筆資料
df.head(3)
```

```python
#學習操作panda表格之，只看特定資料
df[df["產業"]=="半導體"]
```

```python
#學習操作panda表格之，排序(從大到小)
df.sort_values("收盤價",ascending=False)
```

```python
#學習操作panda表格之，只看特定資料之特定欄位
df[["公司名稱","收盤價"]]
```

```python
#學習操作panda表格之，簡單的統計相關公式
print("平均收盤價：", df["收盤價"].mean())
print("最高收盤價：", df["收盤價"].max())
print("最低收盤價：", df["收盤價"].min())
print("總成交量：", df["成交量(張)"].sum())
```

```python
#學習操作panda表格之，分組統計 groupby:按產業分組，計算每個產業的平均收盤價
df.groupby("產業")["收盤價"].mean()
```

```python
#學習操作panda表格之，新增欄位：每張收盤價（收盤價 / 1000 股
df["每張價值(元)"]=df["收盤價"]*1000
df
```

```python
#學習操作panda表格之，把我的表格存成 CSV 檔
df.to_csv("practice_data/sample_stock.csv",index=False,encoding="utf-8-sig")
print("File saved!")
```

```python
import requests
```

```python
# 用一個免費的公開 API 來練習
response=requests.get("https://api.exchangerate-api.com/v4/latest/USD")
print("狀態碼",response.status_code)
print("資料類型",type(response.json()))
```

```python
# 取出免費api的匯率資料
data=response.json()
twd_rate=data["rates"]["TWD"]
jpy_rate=data["rates"]["JPY"]

print("1美元=",twd_rate,"台幣")
print("1美元=",jpy_rate,"日圓")
```

```python
#把匯率API資料轉成 DataFrame
exchange_data=response.json()
rates_df=pd.DataFrame(
    list(exchange_data["rates"].items()),
    columns=["貨幣","匯率"]
)
rates_df.head(10)
```

```python
# 練習操作表格
twd=rates_df[rates_df["貨幣"]=="TWD"]
print("台幣匯率",twd["匯率"].values[0])
print("\n匯率最高的5個貨幣:")
rates_df.sort_values("匯率",ascending=False).head(5)
```

```python
# 進階練習，從匯率表中篩選出亞洲主要貨幣
#ac=asain currencies
ac=["TWD","JPY","KRW","CNY","HKD","SGD"]
asian_rates=rates_df[rates_df["貨幣"].isin(ac)].sort_values("匯率",ascending=False)
asian_rates
```

```python
# 綜合練習把 Pandas 和 API 串在一起：幫台股資料加上美元計價
twd_rate=rates_df[rates_df["貨幣"]=="TWD"]["匯率"].values[0]
df["美元計價"]=(df["收盤價"]/twd_rate).round(2)
df[["公司名稱","收盤價","美元計價"]]
```
